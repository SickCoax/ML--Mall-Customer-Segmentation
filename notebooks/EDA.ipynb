# EDA

```python
# Imports

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Data Load

df = pd.read_csv(r"..\dataset\Mall_Customers.csv")

df
```

```python
# Stats

print(df.describe())
```

```python
# Null Values

print(df.isnull().sum())
```

```python
# Gender Distribution

sns.countplot(
    x = "Gender" ,
    data = df
)
plt.title("Gender Distribution")
plt.grid()
plt.plot()
```

```python
# Income vs Spending Score

sns.scatterplot(
    x = "Annual Income (k$)" ,
    y = "Spending Score (1-100)" ,
    data = df
)
plt.title("Income vs Spending Score")
plt.grid()
plt.show()
```

# # Shows different segments :

# 1. Low Income + Low Spending
# 2. Low Income + High Spending
# 3. Avg Income + Avg Spending
# 4. High Income + Low Spending
# 5. High Income + High Spending

```python
# Age vs Spending Scor

## Scatter Plot
sns.scatterplot(
    x = "Age" ,
    y = "Spending Score (1-100)" ,
    data = df ,
    hue = "Gender"
)
plt.title("Income vs Spending Score")
plt.grid()
plt.show()
```

# The scatter plot suggests a negative relationship between Age and Spending Score. Younger customers tend to spend more, whereas spending scores generally decrease with age also both the gender shows mixed trend
